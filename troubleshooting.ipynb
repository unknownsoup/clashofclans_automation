```python
import subprocess
import os
from datetime import datetime

DEVICE_ID = "R52T501DACZ"
PROJECT_FOLDER = "/home/braedon/Projects/clashbot/attacking/battlePics"  # adjust to your actual project path

def take_screenshot(filename=None):
    if filename is None:
        filename = f"screen_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
    
    save_path = os.path.join(PROJECT_FOLDER, filename)
    print(save_path)
    result = subprocess.run(
        ["adb", "-s", DEVICE_ID, "exec-out", "screencap", "-p"],
        capture_output=True
    )
    
    with open(save_path, "wb") as f:
        f.write(result.stdout)
    
    print(f"Screenshot saved to {save_path}")
    return save_path

# save with auto timestamp
#take_screenshot()

# or save with a specific name
take_screenshot("homebase.png")    
```
