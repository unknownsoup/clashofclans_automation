```python
import subprocess
import os
from datetime import datetime

DEVICE_ID = "R52T501DACZ"
PROJECT_FOLDER = "/home/braedon/Projects/clashbot/attacking/battlePics"  # adjust to your actual project path

def take_screenshot(filename=None):
    if filename is None:
        filename = f"screen_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
    
    save_path = os.path.join(PROJECT_FOLDER, filename)
    print(save_path)
    result = subprocess.run(
        ["adb", "-s", DEVICE_ID, "exec-out", "screencap", "-p"],
        capture_output=True
    )
    
    with open(save_path, "wb") as f:
        f.write(result.stdout)
    
    print(f"Screenshot saved to {save_path}")
    return save_path

# save with auto timestamp
#take_screenshot()

# or save with a specific name
take_screenshot("homebase.png")    
```

```python
import cv2
from tools.simpleActions import Actions

def mark_positions(path, positions, color=(0, 0, 255), radius=2):
    src = cv2.imread(path)
    marked = src
    for x, y in positions:
        marked = cv2.circle(src, center=(x, y), radius=radius, color=color, thickness=3)

    cv2.imshow("Circle Drawn", marked)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# usage
screen = '/home/braedon/Projects/clashbot/screenshots/battlePics/preAttack1.png'
positions = [(0, 250), (314, 310)]
mark_positions(path=screen, positions=positions)
```

```python
from tools.simpleActions import Actions
import cv2

def clean_ocr_number(raw_text):
        # remove all periods
        no_periods = raw_text.replace('.', '')
        
        # split into chunks by whitespace
        chunks = no_periods.split()
        
        # keep only numeric chunks
        numeric_chunks = [c for c in chunks if c.isdigit()]
        
        if not numeric_chunks:
            return 0
        
        # return the longest one as an int
        longest = max(numeric_chunks, key=len)
        return int(longest)

path = "/home/braedon/Projects/clashbot/screenshots/battlePics/preAttack5.png"
screen = cv2.imread(path)
gold = Actions.read_text_from_region(screenshot=screen, x1=0, y1=183, x2=314, y2=228)
elixir = Actions.read_text_from_region(screenshot=screen, x1=0, y1=250, x2=314, y2=310)
dark = Actions.read_text_from_region(screenshot=screen, x1=0, y1=323, x2=314, y2=368)
gold = clean_ocr_number(gold)
elixir = clean_ocr_number(elixir)
dark = clean_ocr_number(dark)


print(f"gold:{gold} elixir:{elixir} dark:{dark}")

        
```
